```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))

import src.datosCocineros

cocineros = ["Valentina", "Mateo", "Camila", "Santiago", "Lucía"]
datos = {}

for d in cocineros:
    datos[d] = {
    "puntaje" : 0 ,
    "mejor ronda" : 0 ,
    "victorias" : 0 
    }

#imprimimos el nombre de la categoria
for ronda in src.datosCocineros.rounds:
    prom=-1
    nom = ""
    print(f" === {ronda['theme']} ===")
    print ("Tabla de posiciones")

    #ponemos dos variable nombre y nota para guardar los datos que necesitamos
    for nombre, nota in ronda['scores'].items():
        suma_notas= sum(nota.values())
        cantidad_jueces = len(nota)
        promedio = suma_notas / cantidad_jueces

        #comprueba el puntaje para determinar el ganador de la ronda
        if (prom < promedio):
            prom= promedio
            nom= nombre

        #suma al puntaje de los jugadores la suma de su nota de esa ronda
        datos[nombre]["puntaje"] += suma_notas

        if (suma_notas > datos[nombre]["mejor ronda"]):
            datos[nombre]["mejor ronda"]= suma_notas

        print(f" {nombre}: {nota}")
    
    print("-____")
    print(f"Ganador: {nom} con {prom:.2f}")

    
    datos[nom]["victorias"] += 1

    print("-" *20)
    
print ("\nTabla de posiciones final: ")

print (f"{'Cocinero':14} {'Puntaje':9} {'Rondas ganadas':12} {'Mejor ronda':12} {'Promedio':8}")
print ("-" * 70)
for nom, info in datos.items():
    promedio_final = info ["puntaje"] / 5
    print (f"{nom:15}{info['puntaje']:6}{info['victorias']:12}{info['mejor ronda']:12}{promedio_final:15.2f}")

```
